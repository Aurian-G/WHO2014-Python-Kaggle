```python
#Filter WHO Data to 2014 only
WHO_2014 = WHO_data[WHO_data["Year"] == 2014]
```

```python
#Create new DF without Country, Year and Status to check out correlation amongest predictors
WHO_2014_filtered = WHO_2014.drop(["Country", "Year", "Status"], axis=1)
```

```python
#Check new DF and make sure columns are removed
WHO_2014_filtered.info()
```

```python
#Correlation Matrix
#Keep correlations that show 0.5 variability or higher with response
#Adult Mortality, Alcohol, BMI (0.48), HIV/Aids, Income Comp, Schooling
corr = WHO_2014_filtered.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
#Create scatterplot matrix
from pandas.plotting import scatter_matrix
scatter_matrix(WHO_2014_filtered[['Life expectancy', 'Adult Mortality', 'Alcohol', ' HIV/AIDS', ' BMI ', 'Income composition of resources', 'Schooling']], 
               alpha = 0.2, figsize = (12, 15), c = 'red', s = 100, diagonal = 'hist')
```

```python
##Assertions made from plots above
#We see Schooling is correlated heavily with income.composition
#Adult Mortality is corr with HIV.AIDS, Income.comp, and Schooling
#Alcohol is corr with Income.Comp and Schooling
#BMI Corr with Income.comp and Schooling
#Based of this analysis, we would want to fit a model without Income.comp because
#it is highly correlated with a lot of the other predictors
#Would also be smart to fit a model without Schooling and keep Income.comp instead
#to assess the difference of fits
#BMI shows a quadratic fit - since we are building interpretation based models, we can leave this predictor out
```

```python
#We want to individually plot BMI to see if there is evidence of a quadratic relationship with life expectancy
#We see evidence of curvature and will move forward without it
sns.set(rc={'figure.figsize':(11.7,8.27)})
bmi_scatter = sns.scatterplot(x=" BMI ", y="Life expectancy", data=WHO_2014_filtered)
bmi_scatter.set_title('Scatterplot of BMI vs Life Expectancy')
```

```python
#Log transform Alcohol and HIV colums to see if variation with Life Expectancy increases
WHO_2014_filtered['LogAlcohol'] = np.log(WHO_2014_filtered['Alcohol'])
WHO_2014_filtered['LogHIV/AIDS'] = np.log(WHO_2014_filtered[' HIV/AIDS'])
```

```python
#Check data frame to make sure variables got added
WHO_2014_filtered.info()
```

```python
#Check correlations
#Alcohol increased from 0.52 to 0.57
#HIV/AIDS increased from -0.61 to -0.78
corr = WHO_2014_filtered[['Life expectancy','LogAlcohol','LogHIV/AIDS']].corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
#Check scatterplot matrix distribution of newly transformed log variables
#We are confident to move forward with the log transformation of these 2 variables to predict Life Expectancy
scatter_matrix(WHO_2014_filtered[['Life expectancy', 'LogAlcohol', 'LogHIV/AIDS']], 
               alpha = 0.2, figsize = (12, 15), c = 'red', s = 500, diagonal = 'hist')
```

```python
#Create new DF only with variables needed for MLR
newdf = WHO_2014.drop(["Country", "Year"], axis=1)
newdf['LogAlcohol'] = np.log(newdf['Alcohol'])
newdf['LogHIV/AIDS'] = np.log(newdf[' HIV/AIDS'])
newdf = newdf[['Life expectancy ', 'Status', 'Adult Mortality', 'LogAlcohol', 'LogHIV/AIDS', 'Income composition of resources', 'Schooling']]

#Filter NA's
newdf = newdf[newdf['Income composition of resources'].notna()]
newdf = newdf[newdf['Schooling'].notna()]
newdf = newdf[newdf['LogAlcohol'].notna()]

#Change name of column LE to get rid of space
newdf.rename(columns={'Life expectancy ':'LifeExpectancy', 'LogHIV/AIDS': 'LogHIVAIDS', 'Adult Mortality':'AdultMortality', 'Income composition of resources':'IncomeCompOfRes'}, inplace=True)

#Check if NA's are gone
newdf.head()
```

```python
newdf.isnull().sum()
```

```python
import statsmodels.api  as sm
import statsmodels.formula.api as smf
```

```python
#Fit our full model
mod = smf.ols(formula='LifeExpectancy ~ C(Status) + AdultMortality + LogAlcohol + \
LogHIVAIDS + IncomeCompOfRes + Schooling', data=newdf).fit()
print(mod.summary())
```
