```python
#Create new DF only with variables needed for MLR
newdf = WHO_2014.drop(["Country", "Year"], axis=1)
newdf['LogAlcohol'] = np.log(newdf['Alcohol'])
newdf['LogHIV/AIDS'] = np.log(newdf[' HIV/AIDS'])
newdf = newdf[['Life expectancy ', 'Status', 'Adult Mortality', 'LogAlcohol', 'LogHIV/AIDS', 'Income composition of resources', 'Schooling']]

#Filter NA's
newdf = newdf[newdf['Income composition of resources'].notna()]
newdf = newdf[newdf['Schooling'].notna()]
newdf = newdf[newdf['LogAlcohol'].notna()]

#Change name of column LE to get rid of space
newdf.rename(columns={'Life expectancy ':'LifeExpectancy', 'LogHIV/AIDS': 'LogHIVAIDS', 'Adult Mortality':'AdultMortality', 'Income composition of resources':'IncomeCompOfRes'}, inplace=True)

#Check if NA's are gone
newdf.head()
```

```python
from sklearn.model_selection import train_test_split

X = newdf.iloc[:,1:6]
y = newdf.iloc[:,0]

dummy = pd.get_dummies(X['Status'],drop_first=True)
X = X.drop('Status', axis=1)
X = pd.concat([X,dummy],axis=1)
```

```python
#Split dataset into train and test split
from sklearn.linear_model import LinearRegression
```

```python
#Run 200 iterations of different train/test splits
RMSElist = []

for i in range(0,200):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
    myreg = LinearRegression()
    myreg.fit(X_train, y_train)
    y_pred = myreg.predict(X_test)
    
    myresiduals = y_test - y_pred
    Rmse = round(np.sqrt(np.mean(myresiduals**2)),2)
    
    RMSElist.append(Rmse)
```

```python
#Average out the 200 iterations and compare with base RMSE
np.mean(RMSElist)
```
